Write program to demonstrate the following aspects of signal on sound/image data.

    -2.Template matching



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# Read the main image
img_rgb = cv2.imread('messi.png')
```

```python
# Convert it to grayscale
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_gray)
# Show greayscale messi photo
```

```python
# Read the template
template = cv2.imread('template.jpg',0)

#show Template 
plt.imshow(template)
```

```python
# Store width and height of template in w and h
w, h = template.shape[::-1]
res = cv2.matchTemplate(img_gray,template,cv2.TM_CCOEFF_NORMED)
```

```python
# Specify a threshold
# Store the coordinates of matched area in a numpy array
threshold = 0.8
loc = np.where( res >= threshold)
```

```python
# Draw a rectangle around the matched region.
for pt in zip(*loc[::-1]):
    cv2.rectangle(img_rgb, pt, (pt[0] + w, pt[1] + h), (0,0,255), 2)
```

```python
# Show the final image with the matched area.
cv2.imwrite('res.png',img_rgb)
```

```python
#show template matched in original image in red box
img = mpimg.imread('res.png')
imgplot = plt.imshow(img)
plt.show()
```
