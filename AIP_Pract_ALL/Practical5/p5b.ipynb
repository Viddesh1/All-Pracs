Write a program to apply various enhancements on images
using image derivatives by implementing 

    -Laplacian operations.

```python
# importing required libraries
import mahotas
import numpy as np
from pylab import gray, imshow, show
import os
import matplotlib.pyplot as plt
import cv2
```

```python
# loading image
img = mahotas.imread('lenna_original.jpg')

# filtering image
img = img[:, :]   # done because of mahotas lib
```

```python
#laplacian operator

# [variables]
# Declare the variables we are going to use
ddepth = cv2.CV_16S
kernel_size = 3
# [variables]
# [load]
src =cv2.imread("lenna_original.jpg",1)
# [load]
# [reduce_noise]
# Remove noise by blurring with a Gaussian filter
src = cv2.GaussianBlur(src, (3, 3), 0)
# [reduce_noise]
# [convert_to_gray]
# Convert the image to grayscale
src_gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
# [convert_to_gray]
# [laplacian]
# Apply Laplace function
dst = cv2.Laplacian(src_gray, ddepth, ksize=kernel_size)
# [laplacian]
# [convert]
# converting back to uint8
abs_dst = cv2.convertScaleAbs(dst)
# [convert]
# [display]

print(" Original Image")

# showing image
imshow(img)
show()

print("When apply laplacian operation")
plt.imshow(abs_dst)
```
