    -Contrast adjustments

```python
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt
```

```python
#read the image
im = Image.open("lenna_original.jpg")
```

```python
#image brightness enhancer
enhancer = ImageEnhance.Contrast(im)
```

```python
# set factors for contrast adjustment
factors = [0.5,1.0,1.5,2.0]

# show different contrast settings of lenna original photo
for i in factors:
    im_output = enhancer.enhance(i)
    imgplot = plt.imshow(im_output)
    plt.show()
```
